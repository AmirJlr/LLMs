```python
from langchain_huggingface import HuggingFaceEmbeddings
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

embedding_name = 'HooshvareLab/bert-fa-base-uncased'
embeddings = HuggingFaceEmbeddings(model_name=embedding_name)

texts = [
    "هوش مصنوعی در حال تغییر جهان است.",
    "فناوری هوش مصنوعی زندگی انسان‌ها را دگرگون می‌کند.",
    "رانندگی علی خوب است"
]

vectors = [embeddings.embed_query(text) for text in texts]
vectors_np = np.array(vectors)

cos_sim_matrix = cosine_similarity(vectors_np)
print(cos_sim_matrix)
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

texts = [
    "هوش مصنوعی در حال تغییر جهان است.",
    "فناوری هوش مصنوعی زندگی انسان‌ها را دگرگون می‌کند.",
    "رانندگی علی خوب است"
]

vectors = model.encode(texts, convert_to_numpy=True)
cos_sim_matrix = util.cos_sim(vectors, vectors).numpy()

print(cos_sim_matrix)
```
